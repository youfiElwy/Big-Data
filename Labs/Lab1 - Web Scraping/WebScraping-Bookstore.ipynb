```python
# Imports
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# Bookstore's page
cbb_url = 'https://www.bookwormegy.com/search-results?q=Harry+Potter'

# Use requests to retrieve data from a given URL
cbb_response = requests.get(cbb_url)

# Parse the whole HTML page using BeautifulSoup
cbb_soup = BeautifulSoup(cbb_response.content, 'lxml')

# Title of the parsed page
print(cbb_soup.title.text)
# print(cbb_soup.prettify())
```

```python
books = cbb_soup.find('ul', {'class': 'sDS4bMt'})
# print(books.prettify())
```

```python
firstBook = books.find('li', {'class': 'sERPSMg'})
# print(firstBook.prettify())
```

```python
title = firstBook.find('div', {'class': 'sJpr9vC'}).a.text
title
```

```python
price = firstBook.find('span', {'class': 'sQq4biu'}).text
price
```

```python
titles=[]
prices=[]

for book in books:

  book_title = book.find('div', {'class': 'sJpr9vC'}).a.text
  book_price = book.find('span', {'class': 'sQq4biu'}).text

  titles.append(book_title)
  prices.append(book_price)

df = pd.DataFrame({'Book Title':titles,'Price':prices})
df
```
